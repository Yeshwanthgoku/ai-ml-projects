```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


months = np.arange(0, 24)
weights = 70 + 0.2*months + np.random.normal(scale=1.2, size=months.shape)

df = pd.DataFrame({'month': months, 'weight': weights})
plt.scatter(df['month'], df['weight']); plt.xlabel('Month'); plt.ylabel('Weight (kg)'); plt.title('Weight trend'); plt.show()

X = df[['month']]
y = df['weight']
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
print("Coef:", model.coef_, "Intercept:", model.intercept_)
print("MSE:", mean_squared_error(y, y_pred), "R2:", r2_score(y, y_pred))

plt.plot(df['month'], y_pred, label='Predicted')
plt.scatter(df['month'], y, label='Actual')
plt.legend(); plt.show()
```
