```python

import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

wine = load_wine()
X = pd.DataFrame(wine.data, columns=wine.feature_names)
y = wine.target

sc = StandardScaler()
X_scaled = sc.fit_transform(X)

pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)
print("Number of components for 95% variance:", pca.n_components_)

plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of components'); plt.ylabel('Cumulative explained variance'); plt.grid(); plt.show()

pca2 = PCA(n_components=2)
X_pca2 = pca2.fit_transform(X_scaled)
df_pca2 = pd.DataFrame(X_pca2, columns=['PC1','PC2'])
df_pca2['target'] = y
sns.scatterplot(data=df_pca2, x='PC1', y='PC2', hue='target', palette='deep')
plt.title('Wine PCA (2 components)'); plt.show()
```
